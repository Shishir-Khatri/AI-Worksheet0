```python
# Task 4. Analyze Average Time Use

# Daily time (in hours): [study, entertainment, sleep]
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]

study_hours = []
entertainment_hours = []
sleep_hours = []

for study, entertainment, sleep in time_data:
    study_hours.append(study)
    entertainment_hours.append(entertainment)
    sleep_hours.append(sleep)

total_days = len(time_data)

avg_study = sum(study_hours) / total_days
avg_entertainment = sum(entertainment_hours) / total_days
avg_sleep = sum(sleep_hours) / total_days

print(f"Average hours spent studying: {avg_study:.2f} hours")
print(f"Average hours spent on entertainment: {avg_entertainment:.2f} hours")
print(f"Average hours spent sleeping: {avg_sleep:.2f} hours")
```

```python
# Task 1 - Sum of Nested Lists

def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
total_sum = sum_nested_list(nested_list)

print(f"Nested List: {nested_list}")
print(f"Total Sum: {total_sum}")
```

```python
# Task 2 - Generate All Permutations of a String

def generate_permutations(s):
    if len(s) <= 1:
        return [s]

    permutations = set()

    for i in range(len(s)):
        char = s[i]
        remaining_string = s[:i] + s[i+1:]

        for perm in generate_permutations(remaining_string):
            permutations.add(char + perm)

    return sorted(list(permutations))

permutations_abc = generate_permutations("abc")
permutations_aab = generate_permutations("aab")

print(f"Permutations of 'abc': {permutations_abc}")
print(f"Permutations of 'aab': {permutations_aab}")
```

```python
# Task 3 - Directory Size Calculation

def calculate_directory_size(directory):
    total_size = 0

    for name, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value

    return total_size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

total_size_kb = calculate_directory_size(directory_structure)

print(f"Directory Structure: {directory_structure}")
print(f"Total Directory Size: {total_size_kb} KB")
```

8.2.1 Exercises - Dynamic Programming (DP)

```python
# Task 1 - Coin Change Problem

def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

coins_list = [1, 2, 5]
target_amount = 11
min_c = min_coins(coins_list, target_amount)

test_impossible = min_coins([2], 3)

print(f"Coins: {coins_list}, Amount: {target_amount}, Min Coins: {min_c}")
print(f"Coins: [2], Amount: 3, Min Coins: {test_impossible}")
```

```python
# Task 2 - Longest Common Subsequence (LCS)

def longest_common_subsequence(s1, s2):
    m = len(s1)
    n = len(s2)

    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

s1_test = "abcde"
s2_test = "ace"
lcs_length = longest_common_subsequence(s1_test, s2_test)

print(f"String 1: '{s1_test}', String 2: '{s2_test}', LCS Length: {lcs_length}")
```

```python
# Task 3 - 0/1 Knapsack Problem

def knapsack(weights, values, capacity):
    n = len(weights)

    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        current_weight = weights[i - 1]
        current_value = values[i - 1]

        for w in range(1, capacity + 1):
            if current_weight <= w:
                value_with_item = current_value + dp[i - 1][w - current_weight]
                value_without_item = dp[i - 1][w]

                dp[i][w] = max(value_with_item, value_without_item)
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

w_test = [1, 3, 4, 5]
v_test = [1, 4, 5, 7]
c_test = 7
max_value = knapsack(w_test, v_test, c_test)

print(f"Weights: {w_test}, Values: {v_test}, Capacity: {c_test}, Max Value: {max_value}")
```
